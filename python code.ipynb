```python
# extraction of DATA using function
def slice_data (df, number_of_rows):
    return df.iloc[:number_of_rows]
first_1000 = slice_data(raw_df, 1000)
```

```python
#print length of extracted data
print(len(first_1000))
```

```python
# distribution of channel type
channel_dist = {}

def dist(t):
    for x in t:
        channel_dist[x] = channel_dist.get(x,0) + 1

distrubition = dist(first_1000.channeltype)
```

```python
# display distribution
print(channel_dist)
```

```python
# channel distribution df
channel_dist_df= pd.DataFrame(list(channel_dist.items()),columns = ['ChannelType','Count']) 
channel_dist_df
```
